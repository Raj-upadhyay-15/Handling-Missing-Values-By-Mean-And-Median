```python
null_var = (df.isnull().sum()/len(df)) * 100
null_var
```

```python
drop_col = null_var[null_var > 20] . index
drop_col
```

```python
df2_drop = df.drop(columns=drop_col)
df2_drop.shape
```

```python
# access numeric data set
df3_num = df2_drop.select_dtypes(include=['int64','float64'])
df3_num
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
# jis bhi row me null values he wo show hojayega
```

```python
df3_num.isnull().sum()
```

```python
missing_num = [var for var in df3_num 
               if df3_num[var].isnull().sum()>0]
missing_num
```

```python
plt.figure(figsize=(5,5))
sns.set()
for i , var in enumerate(missing_num):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var] , bins=20 , kde_kws = {'linewidth':5,'color':'#DC143C'})
```

```python
df4_mean = df3_num.fillna(df3_num.mean())
df4_mean.isnull().sum().sum()
# mtlb hmare dataset me koi bhi missing value present nhi he
```

```python
plt.figure(figsize=(5,5))
sns.set()
for i , var in enumerate(missing_num):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var] , bins=20 , kde_kws = {'linewidth':8,'color':'red'} , label='original')
    sns.distplot(df4_mean[var] , bins=20 , kde_kws = {'linewidth':5,'color':'green'} , label='mean')
    plt.legend()
```

```python
df5_median = df3_num.fillna(df3_num.median())
df5_median.isnull().sum().sum()
# mtlb hmare dataset me koi bhi missing value present nhi he
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i , var in enumerate(missing_num):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var] , bins=20 , hist=False , kde_kws = {'linewidth':8,'color':'red'} , label='original')
    sns.distplot(df4_mean[var] , bins=20 , hist=False , kde_kws = {'linewidth':5,'color':'green'} , label='mean')
    sns.distplot(df5_median[var] , bins=20 , hist=False , kde_kws = {'linewidth':3,'color':'yellow'} , label='median')
    plt.legend()
```

```python
df_concat = pd.concat([df3_num[missing_num],df4_mean[missing_num],df5_median[missing_num]] , axis = 1)
```
